```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (20.0,10.0)

#readingdata 
data = pd.read_csv('houserates.csv')


X = data['X2 house age'].values
Y = data['Y house price of unit area'].values

#mean 
x_mean = np.mean(X)
y_mean = np.mean(Y)

m = len(X)

#Calculating b1 & b0

num = 0
denum = 0
for i in range(m):
  num += (X[i] - x_mean)*(Y[i] - y_mean)
  denum += (X[i] - x_mean)**2

b1 = num/denum
b0 = y_mean - (b1*x_mean)

#print coefficients 
print('b1 value is ', b1) 
print('b0 value is ', b0)

#plotting value and regression lines
max_x = np.max(X) + 100
min_x = np.min(X) - 100

#calculating line values x & y 
x = np.linspace(min_x, max_x, 1000) 
y = b0 + b1*x 

#plotting line
plt.plot(x,y,color='#b23c09', label = 'Regression line')
#plotting scatter points
plt.scatter(X,Y,color='#4d93e1',label = 'Scatter Plot')

plt.xlabel('House Age')
plt.ylabel('House Rates per unit area')
plt.legend()
plt.show() 

#Minimizing error by R square 

ss_t =  0 
ss_r = 0
for i in range (m):
  y_pred = b0 + b1*X[i]
  ss_t += (Y[i] - y_mean)**2
  ss_r += (Y[i] - y_pred)**2

r2 = 1 - (ss_r/ss_t)
print('R Square is ', r2 ) 





```
